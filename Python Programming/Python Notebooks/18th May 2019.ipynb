## Problem 2:
#### Problem Statement: 
Plot a line graph on income vs year for the state of arizona

```python
import numpy as np
t=np.arange(1.0,2.0,0.3)
s=1+np.sin(2*np.pi*t)
plt.xlabel('time')
plt.ylabel('voltage')
plt.plot(t,s)
plt.title('Time Machine')
plt.grid(True)
plt.savefig('Figures/TimeVoltage')
plt.show()
```

## Problem 3:
#### Problem Statement:
Line graph for income of Arizona

```python
import pandas as pd

def incomeplot(inc_filename):
    df=readCSV(inc_filename)
    headings=df.columns
    li=df.values[2]
    
    years=[]
    income=[]
    for i in range(2,len(headings)):
        years.append(headings[i])
    for i in range(2,len(li)):
        income.append(li[i])
    plt.xlabel("Years")
    plt.ylabel("Income")
    plt.title("Income of Arizona")
    plt.plot(years,income)
    plt.grid(True)
    plt.show()
def readCSV(inc_filename):
    df=pd.read_csv(inc_filename)
    return df
incomeplot('DataFiles/Income.csv')
```

## Problem 4:
#### Problem Statement:
Generate Scatter plot for 50 different numbers

```python
n=50
x=np.random.rand(n)
y=np.random.rand(n)
colorlist=np.random.rand(n)
arealist=np.pi*(15*np.random.rand(50))**2
plt.scatter(x,y,s=arealist,c=colorlist,alpha=10)
plt.savefig('Figures/Scatter-plot-50-random-circles')
```
